```python
import pandas as pd
import numpy as np
import datetime as dt
import io
df_cases=pd.read_csv('CasosPositivosCOVID-19_7Enero.csv',dtype={ 
                                                                'ID de caso':str,
                                                                'C√≥digo DIVIPOLA departamento':str,
                                                                'Nombre departamento':str,    
                                                                'C√≥digo DIVIPOLA municipio':str,    
                                                                'Nombre municipio':str,         
                                                                'Edad':int,        
                                                                'Unidad de medida de edad':str,     
                                                                'Sexo':str,     
                                                                'C√≥digo ISO del pa√≠s':str,    
                                                                'Nombre del pa√≠s':str,    
                                                                'Recuperado':str,    
                                                                'Tipo de recuperaci√≥n':str,       
                                                                'Pertenencia √©tnica':str,       
                                                                'Nombre del grupo √©tnico':str
                                                                },low_memory=False)
df_cases.head()
```

```python
df_Dept=df_cases[['C√≥digo DIVIPOLA departamento','Nombre departamento','Sexo','Edad','Recuperado']].dropna()
```

```python
def alentados(fila):
    r=fila['Recuperado']
    if r=='Recuperado':
        re=1
        return re
    else:
        re=0
        return re
df_Dept['Alentados']=df_Dept.apply(alentados,axis=1)
```

```python
def activos(fila):
    a=fila['Recuperado']
    if a=='Activo':
        ac=1
        return ac
    else:
        ac=0
        return ac
df_Dept['Activos']=df_Dept.apply(activos,axis=1)
```

```python
def fallecidos(fila):
    f=fila['Recuperado']
    if f=='Fallecido' or f=='fallecido':
        fa=1
        return fa
    else:
        fa=0
        return fa
df_Dept['Fallecidos']=df_Dept.apply(fallecidos,axis=1)
```

```python
def generof(fila):
    sexf=fila['Sexo']
    if sexf =='F':
        resultado=1
        return resultado
    else:
        resultado=0
        return resultado
df_Dept['Femenino']=df_Dept.apply(generof,axis=1)
```

```python
def generom(fila):
    sexm=fila['Sexo']
    if sexm =='M':
        respuesta=1
        return respuesta
    else:
        respuesta=0
        return respuesta
df_Dept['Masculino']=df_Dept.apply(generom,axis=1)
```

```python
def cluster_edad(fila):
    decil=fila['Edad']
    if decil<=9:
        resultado="0-9"
        return resultado
    else:
        if decil>=10 and decil<=19:
            resultado="10-19"
            return resultado
        else:
            if decil>=20 and decil<=29:
                resultado="20-29"
                return resultado
            else:
                if decil>=30 and decil<=39:
                    resultado="30-39"
                    return resultado
                else:
                    if decil>=40 and decil<=49:
                        resultado="40-49"
                        return resultado
                    else:
                        if decil>=50 and decil<=59:
                            resultado="50-59"
                            return resultado
                        else:
                            if decil>=60 and decil<=69:
                                resultado="60-69"
                                return resultado
                            else:
                                if decil>=70 and decil<=79:
                                    resultado="70-79"
                                    return resultado
                                else:
                                    if decil>=80 and decil<=89:
                                        resultado="80-89"
                                        return resultado
                                    else:
                                        if decil>=90:
                                            resultado="90+"
                                            return resultado
df_Dept['Rango_Edad']=df_Dept.apply(cluster_edad,axis=1)
```

```python
df_Report=df_Dept.groupby(by=['C√≥digo DIVIPOLA departamento',
                                   'Nombre departamento',
                                   #'C√≥digo DIVIPOLA municipio',
                                   #'Nombre municipio',
                                   'Rango_Edad',
                                   'Sexo'
                                  ]).agg({'Edad':'mean',
                                          'Recuperado':'count',
                                          'Femenino':'sum',
                                          'Masculino':'sum',
                                          'Alentados':'sum',
                                          'Activos':'sum',
                                          'Fallecidos':'sum'
                                         }).reset_index()
df_Report
```

### se presento la teoria la tasa media de muertes en las mujeres era superior 

```python
def mujeres(fila):
    s=fila['Sexo']
    f=fila['Fallecidos']
    if s=='F':
        return f 
    else:
        return 0

df_Report['MujeresFallecidas']=df_Report.apply(mujeres,axis=1)
```

```python
def Hombres(fila):
    s=fila['Sexo']
    f=fila['Fallecidos']
    if s=='M':
        return f 
    else:
        return 0

df_Report['HombresFallecidos']=df_Report.apply(Hombres,axis=1)
```

```python
#Tasa alentados
def tasa_alentados(fila):
    tasa=(fila['Alentados']/fila['Total'])*100 #+str('%')
    return tasa
df_Report.loc[:,'Tasa_%-Recuperados']=df_Report.apply(tasa_alentados,axis=1)
#total_recuperados
```

```python
#Tasa Activos
def tasa_Activos(fila):
    tasa=(fila['Activos']/fila['Total'])*100 #+str('%')
    return tasa
df_Report.loc[:,'Tasa_%-Activos']=df_Report.apply(tasa_Activos,axis=1)
#total_Activos
```

```python
#Tasa fallecidos
def tasa_fallecidos(fila):
    tasa=(fila['Fallecidos']/fila['Total'])*100 #+str('%')
    return tasa
df_Report.loc[:,'Tasa_%-Decesos']=df_Report.apply(tasa_fallecidos,axis=1)
#total_Fallecidos
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(9,7))
re=df_Report['Tasa_%-Recuperados']
ac=df_Report['Tasa_%-Activos']
de=df_Report['Tasa_%-Decesos']
ed=df_Report['Edad_Prom']
plt.bar(ed,de,color="red",label="Muertos")
plt.bar(ed,ac,color="blue",label="Activos")
plt.bar(ed,re,color="green",bottom=[sum(data) for data in zip(de,ac)],label="Recuperados")
#plt.bar(year,data1,color="red",bottom=[sum(data) for data in zip(data2,data3)],label="C++")

plt.legend(loc="lower left",bbox_to_anchor=(0.05,1.0))
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


labels = df_Report['Edad_Prom']
men_means = df_Report['Masculino']
women_means = df_Report['Femenino']

x = np.arange(len(labels))  # the label locations
width = 0.85  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Cantidad de contagios')
ax.set_title('Edad por Genero')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom',fontsize=6)

autolabel(rects1)
autolabel(rects2)
fig.tight_layout()
plt.show()
```
